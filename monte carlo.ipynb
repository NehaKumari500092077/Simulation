```
from numpy.lib.function_base import interp
import pandas as pd

def calArivalTime(start_time, duration):
  list1 = str(start_time).split(":")
  list1[1:2] = list1[1].split(" ")
  while int(list1[1]) + duration > 59:
    list1[1] = str(int(list1[1]) - 60)
    if int(list1[0]) + 1 > 12:
      list1[0] = '1' 
      list1[2] = 'pm'
    else:
      list1[0] = str(int(list1[0]) + 1) 

  if int(list1[1]) + duration < 60:
    if int(list1[1]) + duration < 10:
      list1[1] = '0' + str(int(list1[1]) + duration)
    else:
      list1[1] = str(int(list1[1]) + duration)

  if int(list1[0]) < 10:
    list1[0] = '0' + str(int(list1[0]))

  return str(list1[0])+":"+str(list1[1])+" "+str(list1[2])

def calWaitingTime(s_start_time, a_time):
  list1 = str(a_time).split(":")
  list1[1:2] = list1[1].split(" ")
  list2 = str(s_start_time).split(":")
  list2[1:2] = list2[1].split(" ")
  count = (int(list1[0]) - int(list2[0]))*60
  count += (int(list1[1]) - int(list2[1]))
  return abs(count)

def calIdleTime(s_end_time, a_time):
  count = int(0)
  list1 = str(s_end_time).split(":")
  list1[1:2] = list1[1].split(" ")
  list2 = str(a_time).split(":")
  list2[1:2] = list2[1].split(" ")
  count = (int(list1[0]) - int(list2[0]))*60
  count += (int(list1[1]) - int(list2[1]))
  if(count < 0):
    return abs(count)
  else:
    return 0

def calServiceStartTime(s_end_time, a_time, duration):
  count = int(0)
  list1 = str(s_end_time).split(":")
  list1[1:2] = list1[1].split(" ")
  list2 = str(a_time).split(":")
  list2[1:2] = list2[1].split(" ")
  count = (int(list1[0]) - int(list2[0]))*60
  count += (int(list1[1]) - int(list2[1]))
  if(count < 0):
    return a_time
  else:
    calArivalTime(s_end_time, duration)


a={'Category': ['Filling', 'Crown', 'Cleaning', 'Extracting', 'Checkup'],
   'Time_required(min)': [45, 60, 15, 45, 15]}
random_value = [40,82,11,34,25,66,17,79]
a['no_of_patient'] = int(0)
a['Probability'] = ' '
a['Cumulative_Prob'] = ' '
a['Range'] = ' '
a['Random_no'] = int(0)
b = pd.DataFrame(a) 
 
Service_hour = int(input("Enter serving hour: "))
single_patient_service_hour = int(input("Dentist gives time to each patient(min) : "))
no_of_patient_in_ser_hr = int((Service_hour * 60)/single_patient_service_hour)
arrival_time = str(input("Enter arrival time in hh:mm am/pm format:  "))
#insert no of patient for each category
index = total_no_of_patient = int(0)
for x in a['Category']:
  no_of_patient = int(input("Enter no of patient for Category "+x+" "))
  total_no_of_patient += no_of_patient
  b.at[index, 'no_of_patient'] = no_of_patient
  index += 1

index = lower_range = int(0)
probability_sum = float(0.00)
#cal probability and comulative pprobability
for x in range(len(a['Category'])):
  b.at[index, 'Probability'] = float(format(float(b.loc[index]['no_of_patient']/total_no_of_patient), '.2f'))
  probability_sum = probability_sum + b.loc[index]['Probability']
  b.at[index, 'Cumulative_Prob'] = float(format(probability_sum, '.2f'))
  #cal range
  b.at[index, 'Range'] = range(lower_range, int(b.loc[index]['Cumulative_Prob'] * 100.0))
  lower_range = int(b.loc[index]['Cumulative_Prob'] * 100)
  index += 1 

dict2 = {'patient': [''], 'Arrival': [' '], 'Service_start': [' '], 'Random_no':[''],
         'service_dur': [''], 'service_end': [''], 'Waiting_time': [' '],
         'Idle_time':[''], 'Category': ['']}

dict2df = pd.DataFrame(dict2)
index = int(0)
for x in range(no_of_patient_in_ser_hr):
  dict2df.at[index, 'patient'] = index + 1
  #print(dict2df.loc[index]['patient'])
  for i in range(len(a['Category'])):
    if(random_value[x] in b.loc[i]['Range']):
      dict2df.at[x, 'Category'] = b.loc[i]['Category']
      break
  index += 1

index = int(0)
dict2df.at[0, 'Arrival'] = arrival_time
for x in dict2df['Category']:
  i = int(b[b['Category'] == x].index.values)
  #print("\nx and i value is: ", x, i)
  dict2df.at[index, 'service_dur'] = b.loc[i]['Time_required(min)']
  dict2df.at[index, 'Random_no'] = random_value[index]
  index += 1
  if(len(dict2df['Category']) != index):
    #print("\n index = ",index)
    dict2df.at[index, 'Arrival'] = calArivalTime(dict2df.loc[index-1]['Arrival'], single_patient_service_hour)
  

dict2df.at[0, 'Service_start'] = dict2df.at[0, 'Arrival']
dict2df.at[0, 'Waiting_time'] = int(0)
dict2df.at[0, 'Idle_time'] = int(0)
for x in range(no_of_patient_in_ser_hr-1):
  dict2df.at[x, 'service_end'] = calArivalTime(dict2df.loc[x]['Service_start'], int(dict2df.loc[x]['service_dur']))
  dict2df.at[x+1, 'Service_start'] = calArivalTime(dict2df.loc[x]['Service_start'], int(dict2df.loc[x]['service_dur']))
 # dict2df.at[x+1, 'Service_start'] = calServiceStartTime(dict2df.loc[x]['service_end'], dict2df.loc[x+1]['Arrival'] ,int(dict2df.loc[x]['service_dur']))
  dict2df.at[x+1, 'Waiting_time'] = calWaitingTime(dict2df.loc[x+1]['Service_start'], dict2df.loc[x+1]['Arrival'])
  dict2df.at[x+1, 'Idle_time'] = calIdleTime(dict2df.loc[x]['service_end'], dict2df.loc[x+1]['Arrival'])

dict2df.at[no_of_patient_in_ser_hr-1, 'service_end'] = calArivalTime(dict2df.loc[no_of_patient_in_ser_hr-1]['Service_start'], 
                                                          int(dict2df.loc[no_of_patient_in_ser_hr-1]['service_dur']))


print("\n\n",b[['Category', 'Time_required(min)', 'no_of_patient']])
print("\n\nTotal no of patient is: ", total_no_of_patient)
print("\n\n",b[['Category','Time_required(min)', 'no_of_patient', 'Probability']])
print("\n\n",b[['Category', 'Probability', 'Cumulative_Prob']])
print("\n\n",b[['Range']])
print("\n\n",dict2df[['patient','Arrival', 'Random_no', 'Category', 'service_dur']])

print("\n\n",dict2df[['patient','Arrival', 'Service_start', 'service_dur', 'service_end','Waiting_time', 'Idle_time']])

avg_waiting_time = avg_idle_time = float(0)
for x in range(no_of_patient_in_ser_hr):
  avg_waiting_time += dict2df.loc[x]['Waiting_time']
  avg_idle_time += dict2df.loc[x]['Idle_time']
print("\n\n Average Waiting Time(min): ", avg_waiting_time/no_of_patient_in_ser_hr)
print("\n\n Average Idle Time(min): ", avg_idle_time/no_of_patient_in_ser_hr)
#print("\n",dict2df[['Category', 'service_dur', 'patient', 'Arrival', 'Service_start', 'Waiting_time']])
```
